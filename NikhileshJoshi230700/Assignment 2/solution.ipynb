```python
file_path = 'cipher.txt'

with open(file_path, 'r') as file:
    hex_data = file.read().strip()

#hex_data = hex_data[2:]
encrypted_message= bytes.fromhex(hex_data)
n = len(encrypted_message)

EnglishAlphabetConstant = float(0.065)
maxKeyLength = int(14)
minKeyLength = int(1) 
def find_key_length() :
    global n 
    global encrypted_message
    ans = int(0) 
    max_constant = float(0.0)

    for key_len in range (minKeyLength , maxKeyLength):
        
            frequency = [int(0)]*256 
            i = int(0) 
            while i*key_len<n :
                frequency[int(encrypted_message[i*key_len])] += 1 
                i += 1

            curr_sum = int(0)
            for f in frequency :
                curr_sum += f*f 


            curr_const = float(curr_sum/(i*i))

            if curr_const > max_constant :
                ans = key_len 
                max_constant = curr_const 
   

    return ans             
                     
def findKey():
    global n 
    global encrypted_message
    key_length = find_key_length()
    english_dictionary  = {97: 8.2, 98: 1.5, 99: 2.8, 100: 4.3, 101: 12.7, 102: 2.2, 103: 2.0, 104: 6.1, 105: 7.0, 106: 0.2, 107: 0.8, 108: 4.0, 109: 2.4, 110: 6.7, 111: 1.5, 112: 1.9, 113: 0.1, 114: 6.0, 115: 6.3, 116: 9.1, 117: 2.8, 118: 1.0, 119: 2.4, 120: 0.2, 121: 2.0, 122: 0.1}
    possible_key = [0]*key_length

    for i in range (key_length) :
        best_key = int(0) 
        max_const = float(0.0)
        for key_character in range(256) :
            shifted_frequency = [0]*256 
            flag = True 
            freq = int(0)
            curr_const= float(0.0)
            j = i
            while j<n : 
                temp = int(encrypted_message[j])^key_character
                if temp not in range(32,128) :
                    flag = False 
                    break 
                if temp in range (48,58) :
                    flag = False 
                    break 
                elif temp in range(65, 91) or temp in range(97, 123):
                    freq += 1
                    shifted_frequency[temp] += 1

                j+= key_length     

            if flag == False :
                continue 

            for test in range(97,123) :
                curr_const += (shifted_frequency[test]*english_dictionary[test])/freq 

            if  curr_const > max_const :
                max_const = curr_const 
                best_key = key_character              



        possible_key[i] = best_key 

    return possible_key  


def encryptVigenere():
    global n
    global encrypted_message
    key = findKey()
    key_length = find_key_length()
    #print(key)
    decrypted_message = []

    for i in range (0,n) :
        decrypted_message.append((encrypted_message[i]^key[i%key_length]))

    return decrypted_message 

decrypted_message = encryptVigenere()

message = ''.join(map(chr, decrypted_message))

with open('plain.txt', 'w') as output_file:
    output_file.write(message)
```
